 # Analysing The Behaviour Of Dataset On The Basis Of Category

## Heatmap of sales in every city by category

```python
data1=data[['City','Category','State','Sales','Profit','Segment']]
```

```python
heatmap1_data=pd.pivot_table(data1,values='Sales',index=['City'],columns='Category')
plt.figure(figsize=(8,12))
sns.heatmap(heatmap1_data,cmap='coolwarm')
```

## Heatmap of sales in every State by category

```python
heatmap2_data=pd.pivot_table(data1,values='Sales',index=['State'],columns='Category')
plt.figure(figsize=(8,12))
sns.heatmap(heatmap2_data,cmap='coolwarm')
```

## Heatmap of Profit in every city by category

```python
heatmap3_data=pd.pivot_table(data1,values='Profit',index=['City'],columns='Category')
plt.figure(figsize=(8,12))
sns.heatmap(heatmap3_data,cmap='BuGn')
```

## Heatmap of Profit in every State by category

```python
heatmap4_data=pd.pivot_table(data1,values='Profit',index=['State'],columns='Category')
plt.figure(figsize=(8,12))
sns.heatmap(heatmap4_data,cmap='BuGn')
```
